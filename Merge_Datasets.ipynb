```python
train=pd.read_csv('train.csv')
features=pd.read_csv('features.csv')
stores=pd.read_csv('stores.csv')
test=pd.read_csv('test.csv')
```

## Merging the three dataframes to create our training dataframe

```python
df_train=train.merge(features, on=['Store','Date'], how='inner').merge(stores,on=['Store'],how='inner')
```

```python
df_train.head()
```

```python
df_test=test.merge(features, on=['Store','Date'], how='inner').merge(stores,on=['Store'],how='inner')
```
